```julia
using PyCall
aseio = pyimport("ase.io")
```

```julia
at1 = aseio.read("run_CH4-model-ANI-test-atomisation/CH4-model-ANI-test-atomisation_atomisation_1.xyz", index=":");
at2 = aseio.read("run_CH4-model-ANI-test-atomisation/CH4-model-ANI-test-atomisation_atomisation_2.xyz", index=":");
at3 = aseio.read("run_CH4-model-ANI-test-atomisation/CH4-model-ANI-test-atomisation_atomisation_3.xyz", index=":");
at4 = aseio.read("run_CH4-model-ANI-test-atomisation/CH4-model-ANI-test-atomisation_atomisation_4.xyz", index=":");
```

```julia
e = hcat([at1[i].info["energy"] for i=1:length(at1)],
    [at2[i].info["energy"] for i=1:length(at2)],
    [at3[i].info["energy"] for i=1:length(at3)],
    [at4[i].info["energy"] for i=1:length(at4)])
```
